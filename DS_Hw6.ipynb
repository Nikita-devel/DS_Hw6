```python
import gdown
import zipfile
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
```

```python
url = 'https://drive.google.com/uc?id=1Zvz20Iqeia1eEtFbGa3NcIrt_SNSimP6'
output = 'dataset.zip'

gdown.download(url, output, quiet=False)

with zipfile.ZipFile(output, 'r') as zip_ref:
    zip_ref.extractall('dataset')

print(os.listdir('dataset'))
```

```python
data_2d = pd.read_csv('/content/dataset/data/data_2d.csv')
```

```python
data_2d
```

```python
X_2d = data_2d.iloc[:, [0, 1]]
scaler_2d = StandardScaler()
X_2d_scaled = scaler_2d.fit_transform(X_2d)
```

```python
model_2d = KMeans()
visualizer_2d = KElbowVisualizer(model_2d, k=(2, 10))
```

```python
visualizer_2d.fit(X_2d_scaled)
visualizer_2d.show()
```

```python
mnist = datasets.fetch_openml('mnist_784')
X_mnist = mnist.data.astype('float32')

scaler_mnist = StandardScaler()
X_mnist_scaled = scaler_mnist.fit_transform(X_mnist)

model_mnist = KMeans()
visualizer_mnist = KElbowVisualizer(model_mnist, k=(2, 10))
```

```python
visualizer_mnist.fit(X_mnist_scaled)
visualizer_mnist.show()
```

```python
from sklearn.decomposition import PCA
```

```python
data_2d = pd.read_csv('/content/dataset/data/data_2d.csv')

print(data_2d.columns)

kmeans_2d = KMeans(n_clusters=2)
data_2d['cluster'] = kmeans_2d.fit_predict(data_2d.iloc[:, [0, 1]])

plt.scatter(data_2d.iloc[:, 0], data_2d.iloc[:, 1], c=data_2d['cluster'], cmap='viridis')
plt.title('K-means Clustering (2D Dataset)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
mnist = datasets.fetch_openml('mnist_784', version=1)
mnist_data = mnist.data.astype(int)

pca = PCA(n_components=2)
mnist_2d = pca.fit_transform(mnist_data)

kmeans_mnist = KMeans(n_clusters=10)
mnist_data['cluster'] = kmeans_mnist.fit_predict(mnist_data)

plt.scatter(mnist_2d[:, 0], mnist_2d[:, 1], c=mnist_data['cluster'], cmap='viridis')
plt.title('K-means Clustering (MNIST Dataset)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
